```python
import numpy as np
import pandas as pd

import nltk
from nltk.corpus import stopwords
import string
```

```python
data = pd.read_csv('Data-Book.csv')
# Keeping only the neccessary columns
data = data[['Sentiment','Text']]
data['Sentiment'] = data['Sentiment'].replace(0, "Negative").replace(1, "Positive")
data.head()
```

```python
data.count()
```

```python
def wordcloud_draw(text):
    #words = ' '.join(data)
    cleaned_word = " ".join([word for word in text.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                            ])
    return cleaned_word

data['cleaned_text'] = data['Text'].apply(lambda x: wordcloud_draw(x))

data.head(20)  
```
