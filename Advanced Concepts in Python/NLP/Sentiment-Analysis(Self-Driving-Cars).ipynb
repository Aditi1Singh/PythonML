```python
import numpy as np
import pandas as pd

import nltk
from nltk.corpus import stopwords
import string
```

```python
data = pd.read_csv('driving-cars-tweets.csv')
# Keeping only the neccessary columns
data = data[['text','sentiment']]
data.head()
```

```python
data.count()
```

```python
def wordcloud_draw(text):
    #words = ' '.join(data)
    cleaned_word = " ".join([word for word in text.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                            ])
    return cleaned_word

data['cleaned_text'] = data['text'].apply(lambda x: wordcloud_draw(x))

data['sentiment'] = data['sentiment'].replace(1, "Negative").replace(2, "Negative").replace(3, "Neutral").replace(4, "Positive").replace(5, "Positive")
data.head(20)  
```

```python
sample = [" ","The food was excellent","The food was EXCELLENT","The food was good but the service was pathetic","10/10 for the ambience","Inspite of the food being good, the manager was very rude to us","The food was pathetic","The food was pathetic!!","I would not eat this food for free also",":-)",":-("]

sample_pd = pd.DataFrame({'text':sample})

from nltk.sentiment.vader import SentimentIntensityAnalyzer

posList = []
negList = []
neuList = []

i = 0

data['prediction'] = ''
data['compound'] = ''

#threshold = 0

sid = SentimentIntensityAnalyzer()
for sentence in data['cleaned_text']:
    vs = sid.polarity_scores(sentence)
    comp = vs['compound']
    if vs['compound'] > 0.65:
        data['prediction'][i] = 'Positive'
        data['compound'][i] = comp
    elif vs['compound'] < -0.65:
        data['prediction'][i] = 'Negative'
        data['compound'][i] = comp
        data['prediction'][i] = 'Negative'
    else:
        data['prediction'][i] = 'Neutral'
        data['compound'][i] = comp
    i = i + 1
    
```
